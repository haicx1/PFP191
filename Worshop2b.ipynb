```python
def s1(n) :
  if n == 1 :
    return 1
  else :
    return (1 + s1(n-1))

def s2(n) :
  if n == 0 :
    return 1
  else :
    return (s2(n-1) * n)

def s3(n) :
  if n == 1 :
    return 1
  else :
    return(s3(n-1) + 1/n)

def isPrimeNumber(n) :
  if n == 1 :
    print(n,'is not a prime number')
  if n >1 :
    for i in range(2,n) :
      if n % i == 0 :
        print(n,'is not a prime number.')
        break
      else :
        print(n,'is a prime number.')

while True :
  try:
    n = int(input('Enter a number:'))
    if n < 5 :
      print('Please re-enter:')
      continue
    else :
      break
  except :
    print('Please re-enter:')

a = s1(n)
b = s2(n)
c = s3(n)
print('S1 =',a)
print('S2 =',b)
print('S3 =',c)
while True :
  try:
    n = int(input('Re-enter n:'))
    break
  except:
    print('Please re-enter:')
isPrimeNumber(n)

```
